## What is Numpy?
NumPy is a Python library used for working with arrays, linear algebra, fourier transform, and matrices.A numpy array is similar to a list. NumPy stands for Numerical Python and it is an open source project.The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.

Arrays are very frequently used in data science, where speed and resources are very important.

NumPy is usually imported under the np alias.

It's usually fixed in size and each element is of the same type. We can cast a list to a numpy array by first importing numpy

```python
## Checking NumPy Version
#The version string is stored under version attribute.
import numpy as np
print(np.__version__)
```

# Sorting arrays

```python
a = np.array([[1, 4, 2],
              [3, 4, 6],
              [0, -1, 5]])

print(a)
```

```python
np.sort(a, axis = None)
```

```python
np.sort(a, axis = 0)
```

```python
values = [('Hrithik', 2009, 8.5), ('Ajay', 2008, 8.7),
           ('Pankaj', 2008, 7.9), ('Aakash', 2009, 9.0)]

dtypes = [('name', 'S10'), ('grad_year', int), ('cgpa', float)]

arr = np.array(values, dtype = dtypes)
print(arr)
```

```python
print(np.sort(arr, order = 'grad_year'))
```

```python
# Write a NumPy program to create an array of all the even integers from 30 to 70.

arr = np.arange(30, 71, 2)
print(arr)
```

```python
# Write a NumPy program to create a 3x4 matrix filled with values from 10 to 21.
arr = np.arange(10, 22).reshape(3, 4)
print(arr)
```

```python
# Write a NumPy program to find the number of rows and columns of a given matrix.

arr = np.array([[1,2,3,7,8,14],
                [4,5,6,10,13,15]])
print('Number of rows : ', arr.shape[0])
print('Number of columns : ', arr.shape[1])
```

```python
# Write a NumPy program to sort a given array by row OR column in ascending order.
arr = np.array([[1,2,3,7,8,14],
                [4,5,6,10,13,15]])
print('Sorting array as per row : ', np.sort(arr, axis = 1))
print('Sorting array as per column : ', np.sort(arr, axis = 0))
```

```python
# Write a NumPy program to swap rows and columns of a given array in reverse order.
arr = np.array([[1,2,3,7,8,14],
                [4,5,6,10,13,15]])

print('Array after swapping rows and columns : ', arr[::-1, ::-1])
```

# Image processing using numpy

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

# PIL - Python Imaging Library is a free and open-source additional library for the Python programming language that adds 
# support for opening, manipulating, and saving many different image file formats. 
# It is available for Windows, Mac OS X and Linux.
```

```python
img = np.array(Image.open('test.jpg'))
plt.figure(figsize = (8,8))
plt.imshow(img)
```

```python
print(img.shape)
print(img.dtype)
print(img.ndim)
```

```python
print(img)
```

```python
plt.figure(figsize = (8,8))
plt.imshow(img[100:200, :200])
```

```python
grayscale_image = img.sum(2)/255*3
print(grayscale_image)

plt.figure(figsize = (8,8))
plt.imshow(grayscale_image)
```

```python
inverted_image = 255 - img

plt.figure(figsize = (8,8))
plt.imshow(inverted_image)
```

```python
state = ['UP', 'AP', 'MP', 'KA', 'TN']
cap = ['LKO', 'AMT', 'BHP', 'BEN', 'CHN']

ser_state_cap = pd.Series(index = state, data = cap)
print(ser_state_cap)
print(ser_state_cap.index)
print(ser_state_cap.values)
print(ser_state_cap['UP'])
print(ser_state_cap[:'MP'])
print(ser_state_cap[['UP', 'MP', 'TN']])
```

```python
# Data Frame - creating data frame

state_df = pd.DataFrame(state, columns = ['State Name'])
state_df
```

```python
cap_df = pd.DataFrame(cap, columns = ['Capitals name'])
cap_df
```

```python
state_cap = list(zip(state, cap))
state_cap
```

```python
state_cap = pd.merge[(state_df, cap_df), axis = 1]
```

```python
state_cap_df = pd.DataFrame(state_cap, columns = ['State Name', 'Capital Name'])
state_cap_df
```
