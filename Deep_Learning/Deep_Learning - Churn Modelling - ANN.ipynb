```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
tf.__version__
```

```python
df = pd.read_csv('~/Downloads/Learnbay Datascience/Churn_Modelling.csv')
df
```

```python
df.info()
```

```python
X = df.iloc[:, 3:13]
y = df.iloc[:, 13]
```

```python
geography = pd.get_dummies(X['Geography'], drop_first = True)
gender = pd.get_dummies(X['Gender'], drop_first = True)
```

```python
X = pd.concat([X, gender, geography], axis = 1)
X
```

```python
X = X.drop(columns = ['Geography', 'Gender'], axis = 1)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_s = ss.fit_transform(X_train)
X_test_s = ss.transform(X_test)
```

```python
# Model creation

model = Sequential([
    Dense(50, activation = 'sigmoid'), # hidden layer1
    Dropout(0.7),
    
    Dense(25, activation = 'relu', kernel_initializer = 'he_normal'), # hidden layer 2
    Dropout(0.8),
    
    Dense(12, activation = 'sigmoid'), # hidden layer3
    Dropout(0.8),
    
    Dense(8, activation = 'relu', kernel_initializer = 'he_normal'), # hidden layer 4
    Dropout(0.8),
    
    Dense(6, activation = 'sigmoid'), # hidden layer 5
    Dropout(0.8),
    
    Dense(1, activation = 'sigmoid') # output layer 
])
```

```python
# compilation of the model 

model.compile(optimizer = 'SGD', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# fitting the model

history = model.fit(X_train_s, y_train, epochs = 10, validation_split = 0.2, batch_size = 32)
```

```python
model.summary()
```

```python
model.predict(X_test_s)
```

```python
y_pred = model.predict(X_test_s)
y_pred = y_pred > 0.5 # converting continuous data to binary by setting a threshold
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
# Let's try another model

# Initializing model
classifier2 = Sequential()

# Adding the input layer and the first hidden layer
classifier2.add(Dense(units = 36, kernel_initializer = 'he_normal', activation = 'relu', input_dim = 11))
classifier2.add(Dropout(0.3))

# Adding second hidden layer
classifier2.add(Dense(units = 20, kernel_initializer = 'he_normal', activation = 'relu'))
classifier2.add(Dropout(0.4))

# Adding third hidden layer
classifier2.add(Dense(units = 15, kernel_initializer = 'he_normal', activation = 'relu'))
classifier2.add(Dropout(0.2))

# Adding output layer
classifier2.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))

# compiling the model
classifier2.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# fitting the model
model_history1 = classifier2.fit(X_train_s, y_train, epochs = 100, validation_split = 0.33, batch_size = 10)
```

```python
# list all data in history
print(model_history1.history.keys())
```

```python
# Summarize history for accuracy

plt.plot(model_history1.history['accuracy'])
plt.plot(model_history1.history['val_accuracy'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

```python
# Summarize history for loss

plt.plot(model_history1.history['loss'])
plt.plot(model_history1.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```
